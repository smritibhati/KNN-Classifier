```python
def iscategorical(col):
    if col in [5,7,8,9,10,11]:
        return True
    return False
```

```python
def findmean(list):
    return list.mean()
```

```python
def findstdev(list):
    return list.std()
```

```python
def attsummary(data):
    mean=[]
    stdev=[]
    for x in range(len(data.columns)):
#         print("mean and stdev of:",data[data.columns[x]])
        mean.append(findmean(data[data.columns[x]]))
        stdev.append(findstdev(data[data.columns[x]]))
    return mean,stdev
```

```python
def summarizedata(data):
    classdata = {}
    classes= data[data.columns[-5]].unique()
    for c in classes:
        classdata[c]=data[data[data.columns[-5]]==c]
#     print(classes)
    summary={}
    mean=[]
    stdev=[]
    for key in classdata:
        mean, stdev = attsummary(classdata[key])
        summary[key]=(mean,stdev)
    
#     print ("_____" ,stdev)
    return summary
```

```python
def prob(x,mean,stdev):
    print("mean",mean)
    print("stdev",stdev)
    exponent = math.exp(-(math.pow(x-mean,2)/(2*math.pow(stdev,2))))
    ans= exponent / (math.sqrt(2*math.pi)*stdev)
    return ans
```

```python
def calculateprobabilities(testrow,summary):
#     print("testrow",testrow)
    probab={}
    for y in range(0,len(testrow)):
        print("WHHHYYYY",y)
        if y==7:
            continue
        for key in summary:
            if key in probab:
                probab[key].append(prob(testrow[y],summary[key][0][y],summary[key][1][y]))
            else:
                probab[key]=[]
                probab[key].append(prob(testrow[y],summary[key][0][y],summary[key][1][y]))

    return probab
```

```python
def multiply(diction):
    prod={}
    for key in diction:
        l=diction[key]
        product=1
        for x in l:
            product*=x
        prod[key]=product
    return prod
```

```python
def multiplyprobab(probab):
    final={}
    
    for key in probab:
        final[key]=multiply(probab[key])
    
    return final
```

```python
def maxprob(final):
    print("final",final)
    maxp=0
    for key in final:
        if final[key]>maxp:
            maxp=final[key]
            maxv=key
    return maxv
```

```python
correct=0
wrong=0
tp=0
tn=0
def naivebayes(data,testdata):
    global correct
    global wrong
    global tp
    global tn
    summary = summarizedata(data)
#     print("summary",summary)
    
    for x in range(len(testdata)):
        row=testdata.iloc[x]
#         print("row", row)
        probab = calculateprobabilities(row,summary)
        probab = multiply(probab)
        predictedlabel= maxprob(probab)
#         print("predictedlabel",predictedlabel)
        if predictedlabel == row[-5]:
            if predictedlabel==1:
                tp+=1
            else:
                tn+=1
            correct+=1
        else:
            wrong+=1
#     print("tp", tp)
    return (correct/(correct+wrong))
```
